```python
import pandas as pd
import requests

def get_pool_data(
        ) -> pd.DataFrame:
    
    response = requests.get('https://yields.llama.fi/pools').json()
        
    # for x in response:
    #     temp = x
        
    # temp_holder = response[temp]
    
    df = [{'chain': x['chain'],
           'project': x['project'],
           'symbol': x['symbol'],
           'tvlUsd': x['tvlUsd'],
           'apyBase': x['apyBase'],
           'apyReward': x['apyReward'],
           'apy': x['apy'],
           'rewardTokens': x['rewardTokens'],
           'pool': x['pool'],
           'apyPct1D': x['apyPct1D'],
           'apyPct7D': x['apyPct7D'],
           'apyPct30D': x['apyPct30D'],
           'stablecoin': x['stablecoin'],
           'ilRisk': x['ilRisk'],
           'exposure': x['exposure'],
           'predictedClass': x['predictions']['predictedClass'],
           'predictedProbability': x['predictions']['predictedProbability'],
           'binnedConfidence': x['predictions']['binnedConfidence'],
           'poolMeta': x['poolMeta'],
           'mu': x['mu'],
           'sigma': x['sigma'],
           'count': x['count'],
           'outlier': x['outlier'],
           'underlyingTokens': x['underlyingTokens'],
           'il7d': x['il7d'],
           'apyBase7d': x['apyBase7d'],
           'apyMean30d': x['apyMean30d'],
           } for x in response['data']]
    
    df = pd.DataFrame(df)
    
    # df = df.set_index('time')
    
    return df
```

```python
def get_chain_data(
        ) -> pd.DataFrame:
    
    response = requests.get('https://api.llama.fi/chains').json()
        
    # for x in response:
    #     temp = x
        
    # temp_holder = response[temp]
    
    df = [{"gecko_id": x['gecko_id'],
           "tvl": x['tvl'],
           "tokenSymbol": x['tokenSymbol'],
           "cmcId": x['cmcId'],
           "name": x['name'],
           "chainId": x['chainId']
           } for x in response]
    
    df = pd.DataFrame(df)
    
    # df = df.set_index('time')
    
    return df
```

```python
def get_protocol_data(
        ) -> pd.DataFrame:
    
    response = requests.get('https://api.llama.fi/protocols').json()
        
    # for x in response:
    #     temp = x
        
    # temp_holder = response[temp]
    
    df = [{"name": x['name'],
           "gecko_id": x["gecko_id"],
           "symbol": x['symbol'],
           "audits": x['audits'],
           "category": x['category'],
           "name": x['name'],
           "chain": x["chain"],
           "tvl": x["tvl"],
           "change_1h": x["change_1h"],
           "change_1d": x["change_1d"],
           "change_7d": x["change_7d"],
           } for x in response]
    
    df = pd.DataFrame(df)
    
    # df = df.set_index('time')
    
    return df
```

```python
protocol_list = list(pool_df['project'].unique())

protocol_tvl_dict = {}
audits_dict = {}
category_dict = {}
chain_type_dict = {}
change_1h_dict = {}
change_1d_dict = {}
change_7d_dict = {}


for dict_name, metric in zip([protocol_tvl_dict, audits_dict, category_dict, chain_type_dict, change_1h_dict, change_1d_dict, change_7d_dict],
                             ['tvl', 'audits', 'category', 'chain', 'change_1h', 'change_1d', 'change_7d']):
  if metric in ['tvl', 'audits', 'change_1h', 'change_1d', 'change_7d']:
    for protocol_name in protocol_list:
      try:
        dict_name[protocol_name] = float(protocol_df.loc[protocol_df['name'] == protocol_name.replace('-', ' ').capitalize()][metric])
      except TypeError:
        try:
          dict_name[protocol_name] = float(protocol_df.loc[protocol_df['name'] == protocol_name.replace('-', ' ').upper()][metric])
        except TypeError:
          try:
            dict_name[protocol_name] = float(protocol_df.loc[protocol_df['name'] == protocol_name.replace('-', ' ').title()][metric])
          except TypeError:
            try:
              dict_name[protocol_name] = float(protocol_df.loc[protocol_df['name'].str.lower() == protocol_name.replace('-', ' ')][metric])
            except TypeError:
              dict_name[protocol_name] = 'fix this one'
  elif metric in ['category', 'chain']:
    for protocol_name in protocol_list:
      try:
        dict_name[protocol_name] = protocol_df.loc[protocol_df['name'] == protocol_name.replace('-', ' ').capitalize()][metric].iat[0]
      except IndexError:
        try:
          dict_name[protocol_name] = protocol_df.loc[protocol_df['name'] == protocol_name.replace('-', ' ').upper()][metric].iat[0]
        except IndexError:
          try:
            dict_name[protocol_name] = protocol_df.loc[protocol_df['name'] == protocol_name.replace('-', ' ').title()][metric].iat[0]
          except IndexError:
            try:
              dict_name[protocol_name] = protocol_df.loc[protocol_df['name'].str.lower() == protocol_name.replace('-', ' ')][metric].iat[0]
            except IndexError:
              dict_name[protocol_name] = 'fix this one'


protocol_tvl_dict 
```
